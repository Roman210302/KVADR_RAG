```python
from ragas.metrics import context_precision, answer_correctness 
from ragas import evaluate
from ragas.llms import llm_factory
from datasets import load_from_disk
import os
```

```python
os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY"
DF_TO_EVALUATE = "local/path" # имя датасета для проверки. должен быть в формате .hf
DF_RESULT = "local/path" # путь для датасета с результатами в формате parquet
METRICS = ['answer_correctness', 'context_precision']
LLM_NAME = "gpt-4o"
```

```python
# загружаем датасет с ответами нашего RAG-пайплайна
test_df_ds = load_from_disk(f"{DF_TO_EVALUATE}.hf")
```

```python
# инициализируем модель оценщик
ragas_llm = llm_factory(model=LLM_NAME)
```

```python
for metric in METRICS:
    globals()[metric].llm = ragas_llm

ragas_result = evaluate(test_df_ds,
    metrics=[answer_correctness, context_precision],
    llm=ragas_llm,
)
```

```python
# итоговые метрики
ragas_result
```

```python
ragas_result.to_pandas()
```

```python
# сохраняем результат оценки
ragas_result.to_pandas().to_parquet(f'{DF_RESULT}.parquet')
```
